```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function and its derivative
def cost_function(x):
    return (x - 3) ** 2 # f(x) = (x - 3)^2 --> 2(x - 3)
```

```python
def gradient(x):
    return 2 * (x - 3) #izvodot go barame 
```

```python
#initialize parameters 

x = np.random.randn() #random initial value 
learning_rate = 0.1 
epochs = 20
```

```python
# store values for visualization 

x_history = [x]
loss_history = [cost_function(x)]
```

```python
# Gradient Descent Loop 

for i in range(epochs):
    
    grad = gradient(x) # compute gradient 
    
    # X1 = X0 - (learning_rate) * (dy / dx (izvodot)) 
    x = x - learning_rate * grad # update weight  - od random point so odredeni cekori sakame da stigneme do lokalen minumum  --> learning rate . Isto taka mnozime i so samiot izvod potocno zamenuvame so random point kaj x 
    
    x_history.append(x)
    loss_history.append(cost_function(x))
    
    print(f'Epoch {i + 1}: x = {x : .4f}, Loss: {cost_function(x): .4f}')
```

```python
plt.figure(figsize=(8, 5))
plt.plot(loss_history, marker="o", linestyle="--") #using circles as a marker and dashed lines to connect them
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Gradient Descent Optimization")
plt.show()
```
